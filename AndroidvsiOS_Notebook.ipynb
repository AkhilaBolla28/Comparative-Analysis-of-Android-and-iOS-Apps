We can see that there are many NaN (missing) values in our dataset, especially in the Rating and rating Count column.

We can see that, almost 7913 of the rows have null values in place of Ratings. Hence we are taking the mean of the Rating column and filling up the null values.

In case of other variables like, Installs, Size etc, the number of null values are too negligible that one can either drop it or fill it. I am just using a fillna command with forward fill to remove null values in the rest of the columns.

```python
#Rating column
df_1['Rating'].fillna(df_1['Rating'].mean(),inplace=True)

#App Name
df_1['App Name'].fillna(method='ffill',inplace=True)

#Size
df_1['Size'].fillna(method='ffill',inplace=True)

#Installs
df_1['Installs'].fillna(method='ffill',inplace=True)
```

```python
#Checking for Null values in the data set
df_1.isnull().sum()
```

```python
# Verifying the Maximum count in Rating attribute

df_1['Rating'].max()
```

```python
# Removing the special characters from the Install column 
df_1['Install1'] = df_1["Installs"].str.replace('+','')
df_1['Install']=df_1['Install1'].str.replace(',','')
df_1.Size = [x.strip().replace('M', '') for x in df_1.Size]
df_1.Size = [x.strip().replace('k', '') for x in df_1.Size]
```

```python
## Changing few categorical to numerical

df_1["Rating"]= df_1.Rating.apply(lambda x: int(x))
df_1["Price"] = df_1.Price.apply(lambda x: float(x))
df_1['Install']= [int(x) if x.isnumeric() else x for x in df_1['Install']]
```

### Visualization for distribution of app ratings for Android and iOS apps in each category


```python
### BAR CHART

android_ratings_by_category = df_3[df_3['Category'].notna()].groupby('Category')['Rating'].apply(list)
ios_ratings_by_category = df_3[df_3['Primary_Genre'].notna()].groupby('Primary_Genre')['Average_User_Rating'].apply(list)

categories = np.sort(df_3['Category'].unique())

# create bar chart for Android ratings by category
fig, ax = plt.subplots(figsize=(10,10))
ax.bar(np.arange(len(categories))-0.2, [np.mean(android_ratings_by_category[cat]) for cat in categories], width=0.4, label='Android', color='tab:blue')
ax.bar(np.arange(len(categories))+0.2, [np.mean(ios_ratings_by_category.get(cat, np.nan)) for cat in categories], width=0.4, label='iOS', color='tab:orange')
ax.set_xticks(np.arange(len(categories)))
ax.set_xticklabels(categories, rotation=90)
ax.set_xlabel('Category')
ax.set_ylabel('Average Rating')
ax.legend()

plt.show()

```

From the above graph, it is clear that, both android and iOS apps in the weather category have received highest ratings when compared with the other categories. iOS has received lowest rating in the productivity category and android has received lowest rating in food & drink category.

```python
### LINE CHART

android_ratings_by_category = df_3[df_3['Category'].notna()].groupby('Category')['Rating'].apply(list)
ios_ratings_by_category = df_3[df_3['Primary_Genre'].notna()].groupby('Primary_Genre')['Average_User_Rating'].apply(list)

categories = np.sort(df_3['Category'].unique())

# create line chart for Android ratings by category
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(categories, [np.mean(android_ratings_by_category[cat]) for cat in categories], label='Android', color='tab:blue')
ax.plot(categories, [np.mean(ios_ratings_by_category.get(cat, np.nan)) for cat in categories], label='iOS', color='tab:orange')
ax.set_xticks(categories)
ax.set_xticklabels(categories, rotation=90)
ax.set_xlabel('Category')
ax.set_ylabel('Average Rating')
ax.legend()

# add grid lines
ax.grid(True)

plt.show()
```

From the above line graph, it appears that the weather category has the highest average rating for both Android and iOS apps, while the productivity category has the lowest average rating for iOS apps and the food & drink category has the lowest average rating for Android apps.

### Distribution of app ratings for free and paid android and iOS apps

```python
### GROUPED BAR CHART

# create a subset of the data for free android apps
free_android = df_3[(df_3['Free_x'] == True)]

# create a subset of the data for paid android apps
paid_android = df_3[(df_3['Free_x'] == False)]

# create a subset of the data for free iOS apps
free_ios = df_3[(df_3['Free_y'] == True)]

# create a subset of the data for paid iOS apps
paid_ios = df_3[(df_3['Free_y'] == False)]

# create a list of the app rating data for each group
android_ratings = [free_android['Rating'], paid_android['Rating']]
ios_ratings = [free_ios['Average_User_Rating'], paid_ios['Average_User_Rating']]

# create a figure with a single subplot
fig, ax = plt.subplots(figsize=(10,15))

# set the width of each bar
bar_width = 0.35

# set the x-axis tick labels
labels = ('Free', 'Paid')
x_ticks = np.arange(len(labels))

# plot the android data
android_bars = ax.bar(x_ticks - bar_width/2, [r.mean() for r in android_ratings], bar_width, label='Android')
for i, rating in enumerate(android_ratings):
    ax.annotate(f"{rating.mean():.2f}", (x_ticks[i]-bar_width/2, rating.mean()+0.05), ha='center')

# plot the ios data
ios_bars = ax.bar(x_ticks + bar_width/2, [r.mean() for r in ios_ratings], bar_width, label='iOS')
for i, rating in enumerate(ios_ratings):
    ax.annotate(f"{rating.mean():.2f}", (x_ticks[i]+bar_width/2, rating.mean()+0.05), ha='center')

# add labels, title, and legend
ax.set_xlabel('App Type')
ax.set_ylabel('Average Rating')
ax.set_title('Distribution of App Ratings for Free and Paid Apps')
ax.set_xticks(x_ticks)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

From the above grouped bar chart, we can say that the mean rating of free iOS apps is higher than free android apps. The mean rating of paid iOS apps is lower than the paid android apps.
